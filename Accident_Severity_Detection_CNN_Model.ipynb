```python
data = data.map(lambda x,y: (x/255, y))
```

```python
len(data)
```

```python
data.as_numpy_iterator().next()
```

```python
train_size = int(len(data) * 0.8)
val_size = int(len(data) * 0.2)
```

```python
train_size
```

```python
val_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
logdir  = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
history = model.fit(train, epochs=10, validation_data=val, callbacks=[tensorboard_callback])
```

```python
plt.figure()
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
fig.suptitle('Loss')
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure()
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
fig.suptitle('Accuracy')
plt.legend(loc="lower right")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
precision,recall, accuracy = Precision(), Recall(), BinaryAccuracy()
```

```python
test = tf.keras.utils.image_dataset_from_directory('C:/Users/annba/Documents/finalcapstonedata/data/test')
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    precision.update_state(y, yhat)
    recall.update_state(y, yhat)
    accuracy.update_state(y, yhat)
```

```python
print(f" Precision : {precision.result()}, \n Recall : { recall.result()}, \n Accuracy : { accuracy.result()}")
```
